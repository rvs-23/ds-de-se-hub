### 7. Scatter plots

#### What is a scatter plot?
A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables. Scatter plots are used to observe relationships between variables.

#### When should we use a scatter plot?
<ul>
    <li>Scatter plotsâ€™ primary uses are to observe and show relationships between two numeric variables. They are used to identify correlational relationships.</li>
</ul>

#### Best practices!

<ol>
    <li>When we have lots of data points to plot, this can run into the issue of overplotting. Overplotting is the case where data points overlap to a degree where we have difficulty seeing relationships between points and variables. It can be difficult to tell how densely-packed data points are when many of them are in a small area.
<br><br>
There are a few common ways to alleviate this issue. One alternative is to sample only a subset of data points: a random selection of points should still give the general idea of the patterns in the full data. We can also change the form of the dots, adding transparency to allow for overlaps to be visible, or reducing point size so that fewer overlaps occur. As a third option, we might even choose a different chart type like the heatmap, where color indicates the number of points in each bin. Heatmaps in this use case are also known as 2-d histograms.</li>
</ol>

```python
x = np.random.randint(1, 10, 15)
y = np.random.randint(1, 10, 15)

plt.scatter(x, y, s=70, c='red', marker='x')        # s controls the size of the dots, m controls what should appear instead of dots(default)
plt.show()
```

#### We can also add edge color to the default marker and add a little transparency.

```python
x = np.random.randint(1, 10, 15)
y = np.random.randint(1, 10, 15)

plt.scatter(x, y, s=70, c='green', edgecolor='black', alpha=0.7, linewidth=1)
plt.show()
```

#### The colors and sizes can be on a per mark basis rather than applying them to all of the marks. Why would we want multiple colors or sizes?  Having the ability to have multiple colors and sizes actually allows us to add additional infromation into our plot. 
Let's say we wanted to add some additional information to our current plot. Let's pretend that a bunch of people have rated something from 1-10 and we want to know their satisfactionn based on their ratings.

```python
ratings = [7, 5, 5, 7, 8, 4, 9, 2, 9, 7, 5, 9, 7, 7, 6]                 # Now these ratings should give us different colors in different dots.

plt.scatter(x, y, c=ratings, alpha=0.8)
plt.show()
```

#### The colors are not that great. We can change that by choosing a suitable cmap. Just like the marker symbol, we have a lot of cmap options which we can use.

```python
plt.figure(figsize=(7,8))
plt.scatter(x, y, c=ratings, edgecolor='black', cmap='Greens', linewidth=1, alpha=0.8)
plt.show()
```

#### We also need to add a colormap legend which can be done as follows

```python
plt.figure(figsize=(9,8))
plt.scatter(x, y, c=ratings, edgecolor='black', cmap='Greens', linewidth=1, alpha=0.8)
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
plt.show()
```

#### Let's read some actual data from a random YouTube video.

```python
yt = pd.read_csv('random_YT.csv')
yt.head()
```

```python
plt.style.use('seaborn')
```

#### Let's check the relation between view_count and likes

```python
plt.figure(figsize=(12,10))
plt.scatter(yt['view_count'], yt['likes'],
            edgecolor='black', linewidth=0.75,
            cmap='Greens', alpha=0.75)
plt.show()
```

#### We see one outlier is messing up with our visualization. Therefore, we can use a log scale.

```python
plt.figure(figsize=(12,10))
plt.scatter(yt['view_count'], yt['likes'],
            edgecolor='black', linewidth=0.75,
            cmap='Greens', alpha=0.75)

plt.xscale('log')
plt.yscale('log')
plt.title('Trending YT videos on a random day')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```

#### Now, let's use the like-dislike ratio to color code the graph.

```python
plt.figure(figsize=(12,10))
plt.scatter(yt['view_count'], yt['likes'],
            edgecolor='black', linewidth=0.75,
            c=yt['ratio'], cmap='summer', alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('like-dislike ratio')

plt.xscale('log')
plt.yscale('log')
plt.title('Trending YT videos on a random day')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```
